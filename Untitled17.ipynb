```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

img = Image.open('monkey.jpg')
w, h = img.size
img = img.resize((w//4, h//4)) # можно не делать .resize(())

# x = np.array(img, dtype=np.float32) # преобразование из PIL в numpy array
# U, S, V = np.linalg.svd(X, full_matrices=False) # разложение SVD
# # обратное преобразование из numpy array в PIL с сохранение изображения на диск
# Image.fromarray(np.asarray(Y_r, dtype=np.uint8)).save(f'{r}.png')

img_np = np.array(img, dtype=np.float32)




plt.imshow(img_np.astype(np.uint8))
plt.title("Оригинал")
plt.axis("off")
plt.show()

print(f"Размер изображения (матрица): {img_np.shape}")


def svd_compress_channel(channel, k):
    U, S, Vt = np.linalg.svd(channel, full_matrices=False)
    U_k = U[:, :k]
    S_k = np.diag(S[:k])
    Vt_k = Vt[:k, :]
    compressed = U_k @ S_k @ Vt_k
    return compressed


def svd_compress_image(img_np, k):
    compressed_img = np.zeros_like(img_np)
    for ch in range(3):  # три канала RGB
        compressed_img[..., ch] = svd_compress_channel(img_np[..., ch], k)
    compressed_img = np.clip(compressed_img, 0, 255)
    return compressed_img.astype(np.uint8)


ks = [1, 2, 10, 30, 100]

plt.figure(figsize=(15, 6))
plt.subplot(1, len(ks)+1, 1)
plt.imshow(img_np.astype(np.uint8))
plt.title("Оригинал")
plt.axis("off")

for i, k in enumerate(ks, start=2):
    compressed = svd_compress_image(img_np, k)
    plt.subplot(1, len(ks)+1, i)
    plt.imshow(compressed)
    plt.title(f"k={k}")
    plt.axis("off")

plt.show()


original_bytes = len(img.tobytes())
print(f"Оригинал: {original_bytes} байт")

for k in ks:
    compressed = svd_compress_image(img_np, k)
    compressed_bytes = compressed.nbytes
    print(f"k={k}: {compressed_bytes} байт")
```

```python

# 1. МНОГОМЕРНОЕ ГАУССОВО РАСПРЕДЕЛЕНИЕ
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# параметры
M = 200
sigma1 = 0.1
sigma2 = 0.5
alpha = np.pi / 4   # угол поворота (45 градусов)

# генерируем данные
x1 = np.random.randn(M, 1) * sigma1
x2 = np.random.randn(M, 1) * sigma2
X = np.concatenate((x1, x2), axis=1)

# поворот
R = np.array([[np.cos(alpha), -np.sin(alpha)],
              [np.sin(alpha),  np.cos(alpha)]])
X_rot = X @ R.T

# матрица ковариации
C = np.cov(X_rot.T)
print("Covariance matrix:\n", C)

# визуализация
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], alpha=0.6)
plt.title("До поворота")
plt.subplot(1,2,2)
plt.scatter(X_rot[:,0], X_rot[:,1], alpha=0.6)
plt.title("После поворота")
plt.show()

# сравнение с готовой функцией
mean = [0,0]
cov = [[sigma1**2, 0], [0, sigma2**2]]
X_mv = np.random.multivariate_normal(mean, cov, M)

plt.scatter(X_mv[:,0], X_mv[:,1], alpha=0.6)
plt.title("multivariate_normal")
plt.show()
```

```python

# 2. ПЛОТНОСТЬ ВЕРОЯТНОСТИ (PDF)
mu = np.mean(X_rot, axis=0)
C = np.cov(X_rot.T)

xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
XX, YY = np.meshgrid(xx, yy)
pp = np.stack((XX.flatten(), YY.flatten()), axis=1)

m = multivariate_normal(mean=mu, cov=C)
ZZ = m.pdf(pp).reshape(XX.shape)

plt.figure(figsize=(6,5))
IMG = plt.pcolor(XX, YY, ZZ, shading='auto')
plt.colorbar(IMG)
CS = plt.contour(XX, YY, ZZ, levels=[0.2,0.4,0.6,0.8], colors='k')
plt.clabel(CS, inline=1, fontsize=10)
plt.scatter(X_rot[:,0], X_rot[:,1], c="red", alpha=0.5)
plt.title("Плотность вероятности + точки")
plt.show()
```

```python

# 3. БИНАРНАЯ КЛАССИФИКАЦИЯ (MAP)
np.random.seed(0)
X0 = np.random.multivariate_normal([0,0], [[0.3,0],[0,0.3]], 100)
X1 = np.random.multivariate_normal([1,1], [[0.3,0],[0,0.3]], 100)
X_bin = np.vstack([X0, X1])
y_bin = np.array([0]*100 + [1]*100)

# априорные вероятности
p0 = np.mean(y_bin==0)
p1 = np.mean(y_bin==1)

m0 = multivariate_normal(mean=np.mean(X0,axis=0), cov=np.cov(X0.T))
m1 = multivariate_normal(mean=np.mean(X1,axis=0), cov=np.cov(X1.T))

xx, yy = np.meshgrid(np.linspace(-2,3,200), np.linspace(-2,3,200))
pp = np.c_[xx.ravel(), yy.ravel()]
Z = (m0.pdf(pp)*p0 - m1.pdf(pp)*p1).reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=20, cmap="coolwarm", alpha=0.6)
plt.contour(xx, yy, Z, levels=[0], colors='k')  # разделяющая кривая
plt.scatter(X0[:,0], X0[:,1], label="Класс 0", alpha=0.7)
plt.scatter(X1[:,0], X1[:,1], label="Класс 1", alpha=0.7)
plt.legend()
plt.title("Бинарная классификация по Байесу")
plt.show()
```

```python

# 4. ЛИНЕЙНЫЙ ДИСКРИМИНАНТНЫЙ АНАЛИЗ (LDA)
from sklearn.base import BaseEstimator

class myLDA(BaseEstimator):
    def __init__(self):
        pass

    def fit(self, X, y):
        self.classes_ = np.unique(y)
        X0 = X[y==self.classes_[0]]
        X1 = X[y==self.classes_[1]]
        self.mu0 = np.mean(X0, axis=0)
        self.mu1 = np.mean(X1, axis=0)
        self.Sigma = np.cov(X.T)  # общая ковариация
        self.Sigma_inv = np.linalg.inv(self.Sigma)

    def predict(self, X):
        g0 = X @ self.Sigma_inv @ self.mu0 - 0.5*self.mu0.T@self.Sigma_inv@self.mu0
        g1 = X @ self.Sigma_inv @ self.mu1 - 0.5*self.mu1.T@self.Sigma_inv@self.mu1
        return np.where(g1>g0, self.classes_[1], self.classes_[0])

lda = myLDA()
lda.fit(X_bin, y_bin)
y_pred = lda.predict(X_bin)

from sklearn.metrics import accuracy_score
print("LDA accuracy:", accuracy_score(y_bin, y_pred))
```

```python
# 5. НАИВНЫЙ БАЙЕС (NB)
class myNB(BaseEstimator):
    def __init__(self):
        pass

    def fit(self, X, y):
        self.classes_ = np.unique(y)
        self.params_ = {}
        for c in self.classes_:
            Xc = X[y==c]
            mu = np.mean(Xc, axis=0)
            var = np.var(Xc, axis=0)
            self.params_[c] = (mu, var, len(Xc)/len(X))

    def predict(self, X):
        preds = []
        for x in X:
            scores = {}
            for c,(mu,var,prior) in self.params_.items():
                likelihood = -0.5*np.sum(np.log(2*np.pi*var) + (x-mu)**2/var)
                scores[c] = likelihood + np.log(prior)
            preds.append(max(scores, key=scores.get))
        return np.array(preds)

nb = myNB()
nb.fit(X_bin, y_bin)
y_pred_nb = nb.predict(X_bin)
print("NB accuracy:", accuracy_score(y_bin, y_pred_nb))
```

```python

# 6. СРАВНЕНИЕ LDA и NB
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score

X, y = make_classification(n_samples=500, n_features=2, n_classes=2,
                           n_redundant=0, n_informative=2, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lda = myLDA()
lda.fit(X_train, y_train)
y_pred_lda = lda.predict(X_test)

nb = myNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

print("LDA metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred_lda))
print("Precision:", precision_score(y_test, y_pred_lda))
print("Recall:", recall_score(y_test, y_pred_lda))

print("\nNB metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Precision:", precision_score(y_test, y_pred_nb))
print("Recall:", recall_score(y_test, y_pred_nb))
```
