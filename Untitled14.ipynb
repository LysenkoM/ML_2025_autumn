```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from collections import Counter
from mnist import load_mnist

train, validation, test = load_mnist()

X_train_full = np.concatenate((train[0], validation[0]), axis=0)
y_train_full = np.concatenate((train[1], validation[1]), axis=0)

X_test, y_test = test
X_train_full = X_train_full.reshape(X_train_full.shape[0], -1).astype(np.float32)
X_test = X_test.reshape(X_test.shape[0], -1).astype(np.float32)

fig, axes = plt.subplots(2, 5, figsize=(10,4))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train_full[i].reshape(28,28), cmap='gray')
    ax.set_title(f"Label: {y_train_full[i]}")
    ax.axis('off')
plt.show()

class my_PCA:
    def __init__(self, n_components=None):
        self.n_components = n_components
        self.components_ = None
        self.mean_ = None
        self.explained_variance_ = None

    def fit(self, X):
        # Центрирование
        self.mean_ = np.mean(X, axis=0)
        Xc = X - self.mean_
        # SVD
        U, S, Vt = np.linalg.svd(Xc, full_matrices=False)
        self.components_ = Vt[:self.n_components]
        self.explained_variance_ = (S[:self.n_components]**2) / (X.shape[0]-1)
        return self

    def transform(self, X):
        Xc = X - self.mean_
        return Xc @ self.components_.T

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

class my_kNN:
    def __init__(self, n_neighbors=5):
        self.n_neighbors = n_neighbors
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = []
        for x in X:

            distances = np.linalg.norm(self.X_train - x, axis=1)
            nearest_idx = np.argsort(distances)[:self.n_neighbors]
            nearest_labels = self.y_train[nearest_idx]
            most_common = Counter(nearest_labels).most_common(1)[0][0]
            y_pred.append(most_common)
        return np.array(y_pred)


from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)

pca = my_PCA(n_components=50)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

# График объяснённой дисперсии
plt.plot(np.cumsum(pca.explained_variance_)/np.sum(pca.explained_variance_))
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.show()

knn = my_kNN(n_neighbors=5)
knn.fit(X_train_pca, y_train)
y_pred = knn.predict(X_val_pca)
print("Validation accuracy:", accuracy_score(y_val, y_pred))

y_test_pred = knn.predict(X_test_pca)
print("Test accuracy:", accuracy_score(y_test, y_test_pred))

pca2 = my_PCA(n_components=2)
X_train_pca2 = pca2.fit_transform(X_train)

plt.figure(figsize=(8,6))
for digit in np.unique(y_train):
    idx = y_train == digit
    plt.scatter(X_train_pca2[idx, 0], X_train_pca2[idx, 1], s=10, alpha=0.5, label=str(digit))
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.title("MNIST в пространстве двух первых главных компонент")
plt.show()

pca_full = my_PCA(n_components=30)
pca_full.fit(X_train)

plt.figure(figsize=(8,6))
plt.plot(np.arange(1, 31), pca_full.explained_variance_, marker='o')
plt.xlabel("Номер компоненты")
plt.ylabel("Собственное значение (дисперсия)")
plt.title("Первые 30 собственных значений")
plt.grid(True)
plt.show()
```
